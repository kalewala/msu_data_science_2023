```python
# const
data_patch = "data/home_credit_bank/application_train_small.csv"
```

```python
# data
df = pd.read_csv(data_patch)
```

Средний доход и максимальная сумма кредита в разрезе F/M и в разрезе уровня образования

```python
df.groupby(['NAME_EDUCATION_TYPE', 'CODE_GENDER']).agg({'AMT_INCOME_TOTAL': ['mean'], 'AMT_CREDIT': ['max']}).unstack()
```

Средний доход, 1 если средний доход более 200000, иначе 0 в разрезе F/M и в разрезе уровня образования

```python
df.groupby(['NAME_EDUCATION_TYPE', 'CODE_GENDER'])['AMT_INCOME_TOTAL'].mean().apply(lambda x: 1 if x>200000 else 0).unstack()
```
